```julia
sciml_logo_url = "https://raw.githubusercontent.com/SciML/SciMLAssets/main/logo/musm_fixes_final/JuliaDiffEq_icon.svg"
```

```julia
sciml_logo = download(sciml_logo_url)
```

```julia
using Luxor, Colors
@draw begin
    # draw book
    Drawing(1_000, 800, "sciml-book-logo.svg")
    origin()
    background("transparent")
    const sciml_blue = "#213055"
    🐢 = Turtle()
    Pencolor(🐢, sciml_blue)
    Penwidth(🐢, 8)
    squircle(Point(0,0), 260, 340; action = :fill, rt = 0.1)
    squircle(Point(15, 18), 260, 340; action = :stroke, rt = 0.1)
    # layer logo
    sciml = readsvg(sciml_logo)
    transform([1.2 0 0 1.2 0 0])
    placeimage(sciml, Point(-120, -120))
    Pencolor(🐢, "#fff")
    fontsize(42)
    fontface("JuliaMono")
    text("SciML Book", Point(0, -160), halign=:center, valign=:center)
    finish()
    preview()
end
```
